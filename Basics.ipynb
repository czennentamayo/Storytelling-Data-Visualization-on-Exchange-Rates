# Storytelling Data Visualization on Exchange Rates

In this project we will use Storytelling Data Visualization to describe the Euro daily exchange rates between 1999 and 2021. We will primarily be focusing on the exchange rates between the the US Dollar and Euro.

Daria Chmkaeva was the person who constructed this dataset and made in available on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020). The European Central Bank is the source of this data. Note that the dataset regulary gets updated, in this projecte we downloaded the data on January 2021.

```python
#Imports our pandas library
import pandas as pd

#Reads our dataset from a CSV file
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

#Displays the first 5 rows of our dataset
exchange_rates.head()
```

```python
#Displays the last 5 rows of our dataset
exchange_rates.tail()
```

```python
exchange_rates.info()
```

By observing our data above, we can see that there is a total **5,699 rows** and **41 columns**.

In some columns like `Bulgarian lev`, `Brazillian real`, and `Chinese yuan renminbi` there are null values in the year 1999, but have vaules in the year 2021. There are columns which are inverse to this like `Slovenian tolar` and `Slovak koruna` which have values in the year 1999, but have null values in the year 2021.

Most columns have the *object* data type excluding `Iceland krona`, `Romanian leu`, and `Turkish lira` which have the *float64* data type.

## Data Cleaning

```python
#Renames the '[US dollar ]' column to 'US_dollar'; Renames the 'Period\Unit' column to 'Time'
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)

#Changes 'Time' data type to datetime
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sorts the of 'Time' in asecending order
exchange_rates.sort_values('Time', inplace=True)

#Resets the index
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar['US_dollar'].value_counts()
```

By observing our data above, we can see that 62 values in the `US_dollar` column has a value of `-`. We will drope the rows that have these values.

```python
#Drops all rows that have '-' as values in the `US_dollar` column
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

#Converts US_dollar to a float type
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)

euro_to_dollar.info()
```

## Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.title('Price of Dollar Overtime')
plt.xlabel('Time')
plt.ylabel('US Dollar')
plt.show()
```

```python
plt.figure(figsize=(9, 6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3, 2, i)
    plt.plot(euro_to_dollar['Time'], 
             euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout()
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

The graph above shows the rolling mean of US dollar in a 30 days rolling window.

## Euro-Dollar Exchange Rate Based on Last 3 Presedential Terms

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
### Adding the FiveThirtyEight style
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(12, 6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
ax4 = plt.subplot(2,1,2)
axes = [ax1, ax2, ax3, ax4]

### Changes to all the subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    
    
### Ax1: Bush
ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'],
                   alpha=0.3)
ax1.text(731581, 1.9, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731350, 1.8, '(2001-2009)', weight='bold',
        alpha=0.3)

### Ax2: Obama
ax2.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'],
                   alpha=0.3)
ax2.text(734500, 1.9, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734350, 1.8, '(2009-2017)', weight='bold',
         alpha=0.3)

### Ax3: Trump
ax3.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'],
                   alpha=0.3)
ax3.text(736750, 1.9, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736650, 1.8, '(2017-2021)', weight='bold',
         alpha=0.3)

### Ax4: Bush-Obama-Trump
ax4.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax4.plot(obama['Time'], obama['rolling_mean'],
        color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'],
        color='#00B2EE')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(730000, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730000, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

plt.show()
```
